```python
import seaborn as sns
sns.violinplot(x='sehir',y='toplam',data=df)
#sehirlere göre kişilerin sahip olduğu cihaz sayısı yığılımı(aykırılığı)
```

```python
sns.lineplot(x='güvenlik',y='toplam',data=df)
#cihazlarında bilgilerinin,kayıtlı olmasının güvenli olduğunu düşünenlerin sahip olduğu cihaz sayıları
```

```python
print(df['isim'].count())
#Veri setindeki kayıt sayısı
```

```python
import numpy as np
columns_array=np.array(df.columns)
columns_series=pd.Series(columns_array,index=[0,1,2,3,4,5,6,7])
print(columns_series.count()-1)#isim nitelik sayılmadığı için eksiltme yaptım
#Veri setindeki nitelik sayısı
```

```python
print("yas "+str(df['yas'].dtype))
print("sehir "+str(df['sehir'].dtype))
print("toplam "+str(df['toplam'].dtype))
print("cihaz "+str(df['cihaz'].dtype))
print("sifre "+str(df['sifre'].dtype))
print("güvenlik "+str(df['güvenlik'].dtype))
print("vakit "+str(df['vakit'].dtype))
#Veri setindeki niteliklerin tipleri
```

```python
df[df.sehir=='Ankara'].groupby(['güvenlik']).size()
#Cihazlarında bilgilerinin kayıtlı olmasını güvenli bulup,bulmayanlara göre grupla ve Ankarada yaşayanların sayısını getir.
```

```python
df[df.cihaz=='Telefon'].groupby('sehir').filter(lambda g: len(g)<4).groupby('sehir').size()
#Cihazlarından en çok telefonu kullananları,sehirlere göre grupla,4'ten küçük olanları getir.
```

```python
df[df.vakit=='3 saat'].groupby('sifre').size()
#Cihazlarında 3 saat geçirenleri,sifre bilgilerini kayıtlı tutup tutmayanlara göre grupla.
```

```python
df[df.güvenlik=='Evet'].groupby('cihaz').size()
#Bilgilerinin cihazlarında kayıtlıyken güvende olduğunu düşünenleri,en çok kullandıkları cihaza göre grupla.
```

```python
print(df.mean())
#Sayısal nitelikler(yas,toplam)için merkezi eğilim ölçüsü-Ortalama
```

```python
print("yas"+str(df['yas'].mode()))
print("toplam"+str(df['toplam'].mode()))
#Sayısal nitelikler(yas,toplam)için merkezi eğilim ölçüsü-Mod Değeri
```

```python
print(df.median())
#Sayısal nitelikler(yas,toplam)için merkezi eğilim ölçüsü-Medyan Değeri
```

```python
df.std()
#Sayısal nitelikler(yas,toplam)için merkezi yayılım ölçüsü-Standart Sapma
```

```python
df.cov()
#Kovaryans negatif,yani yaş ile sahip olunan toplam cihaz arasında negatif bir ilişki vardır.
```

```python
df.corr()
#Korelasyon negatif,yani yaş arttıkça sahip olunan toplam cihaz sayısı azalıyor.
```

```python
corr=df.corr()
sns.heatmap(corr,
           xticklabels=corr.columns.values,
           yticklabels=corr.columns.values)
#Toplam ve yas arasındaki negatif yönlü ilişkinin şiddeti arttıkça renk koyulaşır.
```
